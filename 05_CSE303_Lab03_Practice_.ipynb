```python
def mean(l):
  sum=0
  for i in l:
    sum+=i
  length=len(l)
  return sum/length
```

```python
import math
def Standard_deviation(l):
  s=0
  m=mean(l)
  for i in range(len(l)):
    s+=(l[i]-m)**2
  s/=(len(l)-1)
  return math.sqrt(s)
```

```python
def pearson(x,y):
 xmean=mean(x)
 ymean=mean(y)
 sum=0
 for i in range(len(x)):
  sum+=((x[i]-xmean)*(y[i]-ymean))
 return sum/((len(x)-1)*Standard_deviation(x)*Standard_deviation(y))
```

```python
x=[43,21,25,42,57,59]
y=[99,65,79,75,87,81]
pearson(x,y)
```
